```python
import numpy as mp
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('/creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
credit_card_data.info()
```

```python
credit_card_data.isnull().sum()
```

```python
credit_card_data.dropna(inplace = True)
```

```python
credit_card_data['Class'].value_counts()
```

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit_card_data.groupby('Class').mean()
```

```python
legid_sample = legit.sample(n=163)
```

```python
legid_sample.head()
```

```python
legid_sample.groupby('Class').mean()
```

```python
new_dataframe = pd.concat([legid_sample, fraud], axis = 0)
```

```python
new_dataframe.value_counts('Class')
```

```python
new_dataframe.groupby('Class').mean()
```

```python
X = new_dataframe.drop(columns = 'Class', axis = 1)
Y = new_dataframe['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify=Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print(training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
X_test_accuracyscore = accuracy_score(X_test_prediction, Y_test)
```

```python
print(X_test_accuracyscore)
```
